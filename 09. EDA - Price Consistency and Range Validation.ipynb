```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
from collections import Counter
from itertools import combinations
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
warnings.filterwarnings("ignore")
```

```python
# Load data
df = pd.read_csv("train.csv")
```

```python
# normalize price
df["price"] = df["price"] / 100_000
df["item_price_min"] = df["item_price_min"] / 100_000
df["item_price_max"] = df["item_price_max"] / 100_000
```

```python
keep_columns = [
    "price",
    "item_price_min",
    "item_price_max",
]

df = (
    df[keep_columns]
    .drop_duplicates(keep="first")
    .copy()
)

df["diff_min"] = df["price"] - df["item_price_min"]
df["diff_max"] = df["item_price_max"] - df["price"]

df["diff_min_pct"] = df["diff_min"] / df["price"]
df["diff_max_pct"] = df["diff_max"] / df["price"]
```
